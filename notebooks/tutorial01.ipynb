## Call EPISuite for a list of CAS numbers

```python
from pyepisuite.utils import (
    search_episuite,
    search_episuite_by_cas,
    submit_to_episuite,
    json_to_episuite,
    json_to_ecosar,
)
from pyepisuite.api_client import EpiSuiteAPIClient
from pyepisuite.expdata import (
    HenryData,
    BoilingPointData,
    VaporPressureData,
    MeltingPointData,
    SolubilityData,
)
```

```python
cas_list = ["50-00-0", "100-00-5", "100-02-7"]
ids = search_episuite_by_cas(cas_list)
ids
```

```python
epi_result, ecosar_result = submit_to_episuite(ids)
epi_result
```

```python
epi_result[0].bioconcentration
```

```python
ecosar_result[0]
```

```python
client = EpiSuiteAPIClient()
res = client.submit(cas=ids[1].cas)
res
```
